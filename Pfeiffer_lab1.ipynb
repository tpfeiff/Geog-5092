```python
#Required packages, file and variables
import fiona
import geopandas as gpd
import pandas as pd

path = r"C:\Users\MEng\Documents\CUDenver_classes\(GEOG 5092) GIS Programming and Automation\Labs\Lab1\lab1\lab1.gpkg"
gpkg = fiona.listlayers(r"C:\Users\MEng\Documents\CUDenver_classes\(GEOG 5092) GIS Programming and Automation\Labs\Lab1\lab1\lab1.gpkg")
polygons = []
tables = []
watershd = []

#print gpkg in order to view files in gpkg
print (gpkg)
```

```python
#allocate files to appropriate list


for file in gpkg:
    if file.startswith('soilmu'):
        polygons.append(file)

for file in gpkg:
    if file.startswith('muaggatt'):
        tables.append(file)
        
for file in gpkg:
    if file.startswith('wbdhu8'):
        watershd.append(file)
        
#check to make sure files went into right list
print (polygons)
print (tables)
print (watershd)

#check column names

        
    



```

```python
# Only run if you need to refernce the fields , not a necessary cell to run if you've explored data tables
print (gpd.read_file(path, layer = gpkg[0])) #spatial table
print (gpd.read_file(path, layer = gpkg [10])) #nonspatial table
```

```python
# joing the tables and adding the new field
merged_list = [] #empty list for new joined tables
for spatial in polygons:
    layer_id = spatial[-5:]
    for table in tables:
        if table[-5:] == layer_id:
            p = gpd.read_file(path, layer = spatial)
            t = gpd.read_file(path, layer = table)
            t = t.drop(columns = "geometry")
            join_pt = p.merge(t, left_on = "MUSYM", right_on = "musym")
            join_pt ["mapunit"] = spatial[-5:]
            merged_list.append(join_pt) 
            
            

            
            
            
```

```python
#9 feature classes merged into 1
new_ftrClass = pd.concat(merged_list)
new_ftrClass






```

```python
#watershed boundary file
water_bound = gpd.read_file(path, layer = gpkg[9])
#intersection of watershed boundaries and merged feature class 
watershed_soil = gpd.overlay(new_ftrClass, water_bound, how = "intersection")
watershed_soil

                         
```

```python
#Part 2 finding the number of features in each watershed
final = watershed_soil.groupby(by = "NAME").NAME.count()
for idx, item in final.iteritems():
    print (f' {idx} has {item} feature classes') 
    
#Simply outputping final gives you the answer as well in a nice table





  

             
             
             
    
             
         
         

    


    
             

     
        
```
